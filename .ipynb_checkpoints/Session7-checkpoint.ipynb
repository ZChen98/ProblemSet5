```python
import pandas as pd
import numpy as np
df = pd.read_csv("campus.csv")
df
```

```python
import seaborn as sns
sns.pairplot(df)
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
mod = smf.ols(formula='crime ~ enroll + police + priv', data=df)
res = mod.fit() 
print(res.summary())
```

```python
mod = smf.ols(formula='crime ~ enroll', data=df)
res = mod.fit() 
print(res.summary())
```

R sqaure is always increasing when we put in more variables <br />
The condition number above is likely caused by multicolinearity <br />
Multicolinearity will increase the variance/std err of $\beta$ hence decrease statisitical significance<br />

```python
#Choose between linear or log-linear model
sns.lmplot(x='enroll',y='crime',data=df)
```

The equation for log-lieanr :
$ log(y) = \beta_0 +\beta_1 log(x)=log(e^{\beta_0}x^{\beta_1}) $

it is equivaluent to:
$ y=e^{\beta_0}x^{\beta_1}$


```python
sns.lmplot(x='lenroll',y='lcrime',data=df)
```

Comparing two graphs, it is difficult to say which model is better.<br /> 
Another problem not shown is heteroscedasticity. <br /> 
This problem will make estimator indefficient. Still unbiased though.<br /> 
Inefficient means the sum of sqaure error is not minimized.

Another problem is heteroscedasticity.<br /> 
(standard deviations correlated with x)

```python
from statsmodels.regression.linear_model import GLS
import seaborn as sns
dfy=df['crime']
dfx=sm.add_constant(df['enroll'])
ols_resid = sm.OLS(df['crime'],sm.add_constant(df['enroll'])).fit().resid
logres=np.log(ols_resid**2)
res_fit = sm.OLS(logres, dfx).fit()
sigma=res_fit.predict()
weight=1/np.exp(sigma/2)
#weight=np.diag(np.exp(sigma/2))
#This is one way to do FGLS, you may use it in your project.
#It doesn't make much difference in this data set.
mod = GLS(df['crime'],sm.add_constant(df['enroll']),weight)
res = mod.fit() 
print(res.summary())
#This is also a way to handle data with Autoregressive process 
#eg. A time series data where each dependent variable of this year may have corrlated with its value of last year
```

```python
#mod = smf.ols(formula='crime ~ priv', data=df)
#res = mod.fit() 
#print(res.summary())

#The code above and below do the same thing
mod = sm.OLS(df['crime'],sm.add_constant(df['priv']))
res = mod.fit() 
print(res.summary())
```

From the regression result above, can we says that private school cause less crime?
Endogeneity problem and ommited variable bias

When we propose a model that ignore an variable that is correlated to both dependent and independent variable,<br /> 
the estimator will be baised. We are not going to proof it here but you can see this example.<br /> 
Public school has higher enrollment number and enrollment is correlated with crime number.<br /> 
Then, by omitted enroll number, we will get some wrong estimator.<br /> 
How to avoid or fix these problems? Instrument Variables or 2SLS. <br /> 
IV need to be correlated with X but not correlated with Y<br /> 
How to find instrument variable? Difficult to find and test. (J test/Sarganâ€“Hansen test can only rule out some cases)<br /> 

```python
mod = smf.ols(formula='enroll ~ priv', data=df)
res = mod.fit() 
print(res.summary())
```

```python
mod = smf.ols(formula='crime ~ enroll + priv', data=df)
res = mod.fit() 
print(res.summary())
```

```python
from statsmodels.sandbox.regression.gmm import IV2SLS
dfy=df['crime']
dfx=sm.add_constant(df['priv'])
dfi=df[['police','enroll']]
mod = IV2SLS(dfy,dfx,dfi)
res = mod.fit() 
print(res.summary())
#This is just an sample in case you would like to use in project.
#Since police and enroll is correlated with crime, they are not valid instrument variables.
```

```python
#A logit model
from statsmodels.discrete.discrete_model import Logit
dfy = df['priv']
dfx1 = sm.add_constant(df[['crime','police','enroll']])
dfx2 = sm.add_constant(df[['enroll']])
mod1 = Logit(dfy,dfx1)
res1 = mod1.fit() 
mod2 = Logit(dfy,dfx2)
res2 = mod2.fit() 
print(res1.summary())
print(res2.summary())
#Which one is better? Why?
```

```python
#The next session use the tips dataset
import seaborn as sns
tips = sns.load_dataset("tips")
tips['tip_percent']=tips['tip']/tips['total_bill']
tips['average_bill']=tips['total_bill']/tips['size']
tips['weekends']=[1 if i in ['Sun','Sat'] else 0 for i in tips['day']]
tips['tip_per_guest']=tips['tip']/tips['size']
tips
```

```python
sns.pairplot(tips)
```

```python
#An sample question: does automatic gratuity placed on large dining parties make sense?
mod = smf.ols(formula='tip_percent ~ size', data=tips)
res = mod.fit() 
print(res.summary())
```

```python
#What about absolute value, maybe guest in large party pay less because their bill is larger
mod = smf.ols(formula='tip_per_guest ~ size', data=tips)
res = mod.fit() 
print(res.summary())
```

```python
#Do people be more generous at weekends?
mod = smf.ols(formula='tip_percent ~ weekends', data=tips)
res = mod.fit() 
print(res.summary())
```
